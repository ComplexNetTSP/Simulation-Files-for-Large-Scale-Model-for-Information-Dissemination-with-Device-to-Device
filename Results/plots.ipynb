```python
%pylab inline
import pickle as p
import numpy as np
```

```python
with open('nolatent/1' + '/I.p', 'rb') as fp:
    Inl = p.load(fp)
with open('latent/4' + '/I.p', 'rb') as fp:
    Il_mu2 = p.load(fp)
```

```python
x = np.linspace(1,70,len(Inl))
```

```python
semilogx(x,np.array(Inl)/15686986, 'r')
semilogx(x,np.array(Il_mu2)/15686986, 'b')
xlabel('Time (Days)')
ylabel('Fraction of the population')
```

```python
semilogx(x,cumsum(np.array(Inl)/15686986), 'r')
semilogx(x,cumsum(np.array(Il_mu2)/15686986), 'b')
xlabel('Time (Days)')
ylabel('Fraction of the population')
```

```python
import plotly.plotly as py
from plotly.graph_objs import *

trace1 = Scatter(
    x=x,
    y=np.array(Inl)/15686986
)
trace2 = Scatter(
    x=x,
    y=np.array(Il_mu2)/15686986
)

data = Data([trace1, trace2])

layout = Layout(
    xaxis=XAxis(
        type='log',
        autorange=True
    )
)
fig = Figure(data=data, layout=layout)
unique_url = py.plot(fig, filename = 'basic-line')
```
