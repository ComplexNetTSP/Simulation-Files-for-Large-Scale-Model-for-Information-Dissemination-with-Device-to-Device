```python
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
from matplotlib.ticker import MaxNLocator
my_locator = MaxNLocator(6)

rcParams['axes.labelsize'] = 9
rcParams['xtick.labelsize'] = 9
rcParams['ytick.labelsize'] = 9
rcParams['legend.fontsize'] = 9
#rcParams['font.family'] = 'serif'
#rcParams['font.serif'] = ['Computer Modern Roman']
rcParams['text.usetex'] = True
rcParams['figure.figsize'] = 7.3, 4.2
rcParams['axes.labelweight'] = 'bold'
rcParams['font.weight'] = 'bold'
rcParams['axes.linewidth'] = 0.75    
rcParams['lines.linewidth'] = 2

f, axarr = plt.subplots(2, 2,sharex=True)

f.text(0.05, 0.5, 'Fraction of the population', ha='center', va='center', rotation='vertical')
f.text(0.5, 0.04, 'Time (Days)', ha='center', va='center')
axarr[0,0].semilogx(x,np.array(Inl)/pop, 'r')
axarr[0,0].semilogx(x,np.array(Il_mu2)/pop, 'b')
axarr[0,0].semilogx(x,np.array(Il_mu3)/pop, 'g')
axarr[0,0].set_xlim(0.1,70.0)
axarr[0,0].set_title('Subprefecture 0')
axarr[0,0].yaxis.set_major_locator(my_locator)
#axarr[0,0].xlabel('Time (Days)')
#axarr[0,0].ylabel('Fraction of the population')


y1 = (np.array(Inl)+np.array(Inl_R))/pop
y2 = (np.array(Il_mu2_R)+
      np.array(Il_mu2_ER)+
      np.array(EIl_mu2)+
      np.array(Il_mu2))/pop
y3 = (np.array(EIl_mu3)+
      np.array(Il_mu3)+
      np.array(Il_mu3_ER)+
      np.array(Il_mu3_R))/pop
axarr[1,0].semilogx(x,y1, 'r')
axarr[1,0].semilogx(x,y2, 'b')
axarr[1,0].semilogx(x,y3, 'g')


axarr[0,1].set_title('Subprefecture 59')
axarr[0,1].semilogx(x,np.array(Inl_59)/pop, 'r')
axarr[0,1].semilogx(x,np.array(Il_mu2_59)/pop, 'b')
axarr[0,1].semilogx(x,np.array(Il_mu3_59)/pop, 'g')
setp( axarr[0,1].get_yticklabels(), visible=False)
axarr[0,1].yaxis.set_major_locator(my_locator)
y11 = (np.array(Inl_59)+np.array(Inl_59_R))/15686986
y22 = (np.array(Il_mu2_59_R)+
      np.array(Il_mu2_59_ER)+
      np.array(EIl_mu2_59)+
      np.array(Il_mu2_59))/15686986
y33 = (np.array(EIl_mu3_59)+
      np.array(Il_mu3_59)+
      np.array(Il_mu3_59_ER)+
      np.array(Il_mu3_59_R))/15686986
axarr[1,1].semilogx(x,y11, 'r')
axarr[1,1].semilogx(x,y22, 'b')
axarr[1,1].semilogx(x,y33, 'g')
setp( axarr[1,1].get_yticklabels(), visible=False)

# Zoom in 
axins = zoomed_inset_axes(axarr[1,1], 4, loc=2) # zoom = 6
axins.semilogx(x,y11, 'r')
axins.semilogx(x,y22, 'b')
axins.semilogx(x,y33, 'g')
axins.set_xlim(0.8, 1.5)
axins.set_ylim(0.15, 0.25)

mark_inset(axarr[1,1], axins, loc1=3, loc2=4, fc="none", ec="0.5")
setp(axins.get_yticklabels(), visible=False)

savefig('fig1.pdf')
savefig('fig1.svg')
```
